```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: 나이(age) 구간화 (30대, 40대, 50대 등으로 나누기)
df['age_group'] = pd.cut(df['customer_age'], bins=[20, 30, 40, 50, 60, 70, 80], labels=['20s', '30s', '40s', '50s', '60s', '70s'])

# Step 2: 각 나이 그룹별로 고객 이탈률(%) 계산
churn_rate = df.groupby('age_group')['attrition_flag'].value_counts(normalize=True).unstack() * 100

# Step 3: 히트맵 시각화
plt.figure(figsize=(8, 5))
sns.heatmap(churn_rate, annot=True, fmt=".2f", cmap="Blues", linewidths=0.5)
plt.title("Churn Rate by Age Group")
plt.xlabel("Attrition Status")
plt.ylabel("Age Group")
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: 신용한도를 특정 구간으로 범주화
df['credit_limit_group'] = pd.cut(df['credit_limit'], 
                                  bins=[0, 5000, 10000, 15000, 20000, 30000], 
                                  labels=['0-5K', '5K-10K', '10K-15K', '15K-20K', '20K+'])

# Step 2: 신용한도 그룹별 고객 이탈률(%) 계산
credit_churn_rate = df.groupby('credit_limit_group')['attrition_flag'].value_counts(normalize=True).unstack() * 100

# Step 3: 히트맵 시각화
plt.figure(figsize=(8, 5))
sns.heatmap(credit_churn_rate, annot=True, fmt=".2f", cmap="Reds", linewidths=0.5)
plt.title("Churn Rate by Credit Limit")
plt.xlabel("Attrition Status")
plt.ylabel("Credit Limit Group")
plt.show()
```

```python
# Step 1: 거래 횟수를 특정 구간으로 범주화
df['transaction_group'] = pd.cut(df['total_trans_ct'], 
                                bins=[0, 20, 40, 60, 80, 100, 150], 
                                labels=['0-20', '20-40', '40-60', '60-80', '80-100', '100+'])

# Step 2: 거래 횟수 그룹별 고객 이탈률(%) 계산
trans_churn_rate = df.groupby('transaction_group')['attrition_flag'].value_counts(normalize=True).unstack() * 100

# Step 3: 히트맵 시각화
plt.figure(figsize=(8, 5))
sns.heatmap(trans_churn_rate, annot=True, fmt=".2f", cmap="Blues", linewidths=0.5)
plt.title("Churn Rate by Total Transactions Count")
plt.xlabel("Attrition Status")
plt.ylabel("Transaction Count Group")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# NaN이 있으면 제거
df = df.dropna(subset=['total_trans_amt'])

# 최대 거래 금액 확인
max_amt = df['total_trans_amt'].max()
print(f"Max Transaction Amount: {max_amt}")

# 거래 금액 범위 설정 (최대값을 고려해서 수정)
bins = [0, 500, 1000, 5000, 10000, 20000, max_amt + 1]

# bins 리스트를 오름차순으로 정렬
bins = sorted(bins)

# 확인
print("Sorted bins:", bins)

# 거래 금액 그룹화
df['transaction_amount_group'] = pd.cut(df['total_trans_amt'], bins=bins, labels=['0-500', '500-1K', '1K-5K', '5K-10K', '10K-20K', '20K+'], right=False)

# 이탈율 계산
churn_rate_amount = df.groupby('transaction_amount_group')['attrition_flag'].value_counts(normalize=True).unstack() * 100

# 히트맵 시각화
plt.figure(figsize=(8, 5))
sns.heatmap(churn_rate_amount, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
plt.title("Churn Rate by Transaction Amount")
plt.ylabel("Transaction Amount Group")
plt.xlabel("Attrition Status")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
file_path = "customer_churn.csv"
df = pd.read_csv(file_path)

# 거래 횟수와 거래 금액을 그룹화
transaction_bins = [0, 20, 40, 60, 80, 100, df['total_trans_ct'].max()]
transaction_labels = ['0-20', '20-40', '40-60', '60-80', '80-100', '100+']
df['transaction_count_group'] = pd.cut(df['total_trans_ct'], bins=transaction_bins, labels=transaction_labels, right=False)

# 거래 금액 범위 설정
max_trans_amt = df['total_trans_amt'].max()
print("최대 거래 금액:", max_trans_amt)  # max 값 확인

# 만약 max 값이 너무 작다면 조정
if max_trans_amt < 20000:  
    max_trans_amt = 50000  # 예비 값 설정

amount_bins = [0, 500, 1000, 5000, 10000, 20000, max_trans_amt]
amount_labels = ['0-500', '500-1K', '1K-5K', '5K-10K', '10K-20K', '20K+']
df['transaction_amount_group'] = pd.cut(df['total_trans_amt'], bins=amount_bins, labels=amount_labels, right=False)

# 이탈율 계산
churn_rate_combined = df.groupby(['transaction_count_group', 'transaction_amount_group'])['attrition_flag'].value_counts(normalize=True).unstack() * 100

# 시각화
plt.figure(figsize=(10, 6))
sns.heatmap(churn_rate_combined, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Churn Rate by Transaction Count & Transaction Amount")
plt.xlabel("Attrition Status")
plt.ylabel("Transaction Count & Amount Group")
plt.show()
```
